计数排序:
* 不稳定(不能保证相同数字原有顺序)
* 时间复杂度:O(n+k),k是整数的范围
* 空间复杂度O(k)

```python
def count_sort(lst):
    mmax, mmin = lst[0], lst[0]
    for i in range(1, len(lst)):
        """找到最大最小值"""
        if lst[i] > mmax:
            mmax = lst[i]
        elif lst[i] < mmin:
            mmin = lst[i]

    nums = mmax - mmin + 1
    counts = [0] * nums # 计数列表
    for i in range(len(lst)):
        counts[lst[i] - mmin] = counts[lst[i] - mmin] + 1 # 该计数列表处元素数量+1

    pos = 0
    # 根据计数列表取出对应元素
    for i in range(mmin, mmax+1):
        for j in range(counts[i-mmin]):
            lst[pos] = i
            pos += 1
    return lst
```

```python
array = [5, 2, 8, 3, 9, 0, -1, 3, 6, 11, 10, 11, 2]
count_sort(array)
```
