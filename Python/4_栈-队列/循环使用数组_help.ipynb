&emsp;&emsp;当一个元素退出队列是,可以使用pop(0)而不是pop()从列表中来有意移除第一个元素,一个pop(0)的操作的调用总是处于最坏的情况.
耗时为$\Omega(n)$.可以改进上面的策略,完全避免调用pop(0).可以用一个指代为空的指针代替这个数组中离队的元素,
并且保留一个显示的变量f来储存当前在最前面的元素的索引.这样一个算法对离队操作而言好事为$O(1)$.几次离队操作后,
这个方法可能会导致下图的情景.

&emsp;&emsp;不幸的是,修改后的方法仍然有一个缺点.在一个栈的设计中,列表的长度就是栈的大小(甚至列表底层的储存数组略大).
对于我们正在考虑的队列的设计,情况更糟.例如,建立一个含有相对较少元素的队列时,系统可能让这些元素存储在一个任意大的列表中.
如果不断重复地往一个队列中添加一个新的元素,然后删除另一个,就会发生这样的情况,即随着时间的推移,
底层列表的大小将逐渐增长到$O(m)$其中m值等于自队列创建以来对队列进行追加元素操作的数量总和,而不是当前队列中元素的数量.

&emsp;&emsp;这样的设计会在一些所需队列大小相对稳定却被长时间使用的应用程序中产生不利的影响.例如,
餐厅点餐队列的长度在某一个时刻基本上不可能超过30个,但是在一天(或者一周),队列的总长度将非常大

&emsp;&emsp;为了开发一种更加健壮的队列实现方法,我们让队列的前端趋于右端,并且让队列内的元素在底层数组的尾部"循环".
假定底层数组的长度为固定值N,它比实际队列中的元素数量大.新的元素在当前队列的尾部利用入队列操作进入队列,
逐步将元素从队列的签名插入索引为N-1的位置,然后紧接着时索引为0的位置,接下来是索引为1的位置.如下图所示.

<img src="../../Other/img/基于数组的队列.png">

&emsp;&emsp;实现这种循环的方法并不困难.当从队列中删除一个元素并欲更新前面的索引时,我们可以使用算式f=(f+1)%N进行计算.例如,
14被3整除得到的商为4余数为2.取模操作时处理一个循环数组的理想操作.举一个具体的例子,如果有一个长度为10的列表,并且一个索引为7的首部,
可以通过计算(7+1)%10=8来更新首部.同样,更新索引为8的将会进入索引为9的单元.但是当从索引为9(数组的最后一个单元)处更新时,
需要计算(9+1)%10,其结果为得到索引为0的位置
