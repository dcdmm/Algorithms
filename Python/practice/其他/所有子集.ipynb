```python
import copy

def subsets(lst):
    result = [[]]
    print('init:  ', result)
    for num in lst:
        for element in result[:]:
            x = copy.copy(element) # 也可以像上面一样写为element[:]
            x.append(num)
            result.append(x)
        print("append:" , result)

seq = ['a', 'b', 'c'] # 无重复数字
subsets(seq)
```
